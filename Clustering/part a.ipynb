```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
iris=pd.read_csv('Iris.csv')
iris
X, y = iris.iloc[:,:-1], iris.Species
```

```python

# Normalize features
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_normalized)
```

```python

df = pd.DataFrame(data=X_pca, columns=["PC1", "PC2"])
df["label"] = y
df['label']=df['label'].map({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
df.to_csv("iris_dataset_2D", index=False)
```

```python
kmeans = KMeans(n_clusters=3, max_iter=50, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(X_pca[:, 0], X_pca[:, 1],c=labels, cmap="viridis", s=30)
plt.title("K-Means Clustering (k=3)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.show()
```
