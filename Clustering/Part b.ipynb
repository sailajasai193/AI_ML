```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
#####A
df=pd.read_csv('iris_dataset_2D')
X = df[["PC1", "PC2"]].values
y_true = df["label"].values
```

```python
##B
kmeans = KMeans(n_clusters=3, max_iter=50, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
```

```python
##c
plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="viridis", s=30)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c="red", marker="X", s=200, label="Centers")
plt.title("K-Means Clustering (k=3)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#######D
print(f"Inertia for k=3: {kmeans.inertia_:.2f}")
```

```python
kmeans_1 = KMeans(n_clusters=3, random_state=42)
kmeans_1.fit(X)
cluster_labels_1 = kmeans_1.labels_
df['Cluster_Label_Optimum'] = cluster_labels_1
cluster_label_counts_1 = df.groupby('label')['Cluster_Label_Optimum'].value_counts().unstack(fill_value=0)

for cluster,label_counts in cluster_label_counts_1.items():
    print(f"Cluster {cluster} : {label_counts}")

max_class_count_1 = cluster_label_counts_1.max(axis=0).sum()
N = len(df)
#print(N)
print(f"\nPurity Score is {max_class_count_1/N}")
```

```python
######F
k_values = [2, 4, 6, 10]

plt.figure(figsize=(16, 10))
for i, k in enumerate(k_values, 1):
    km = KMeans(n_clusters=k, max_iter=50, random_state=42)
    km.fit(X)
    preds = km.labels_
    centers = km.cluster_centers_
    
    plt.subplot(2, 2, i)
    plt.scatter(X[:, 0], X[:, 1], c=preds, cmap="Set2", s=30)
    plt.scatter(centers[:, 0], centers[:, 1], c='red', s=150, marker='X')
    plt.title(f"K-Means Clustering (k={k})")
    plt.xlabel("PC1")
    plt.ylabel("PC2")

plt.tight_layout()
plt.show()

inertias = []
k_range = [2, 4, 6, 10]
for k in k_range:
     km = KMeans(n_clusters=k, max_iter=50, random_state=42)
     km.fit(X)
     inertias.append(km.inertia_)
print(inertias)


plt.figure(figsize=(6, 5))
plt.plot(k_range, inertias, 'bo-')
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Curve")
plt.grid(True)
plt.show()

k= KMeans(n_clusters=4, max_iter=50, random_state=42)
k.fit(X)
labels = k.labels_
print(f"Inertia for k=4: {k.inertia_:.2f}")
```
