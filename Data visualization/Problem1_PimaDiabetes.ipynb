```python
#q4
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import load_dataset
df=pd.read_csv('Downloads/pima-indians-diabetes.csv')
mean=df.mean()
print(f"mean of data is {mean}")
median=df.median()
print(f"median of data is {median}")
mode=df.mode()
print(f"mode of data is {mode}")
minimum=df.min()
print(f"minimum of data is {minimum}")
maximum=df.max()
print(f"maximum of data is {maximum}")
st=df.std()
print(f"standard eviation  of data is {st}")
q1=df.quantile(0.25)
print(f"quartile1 of data is {q1}")
q2=df.quantile(0.5)
print(f"quartile2 of data is {q2}")
q3=df.quantile(0.75)
print(f"quartile3 of data is {q3}")
```

```python
import seaborn as sns
from seaborn import load_dataset
datas = pd.read_csv('Downloads/pima-indians-diabetes.csv')
sns.histplot(x=datas['skin'],bins=10, kde=True, data=datas)
plt.xlabel('Skin')
plt.ylabel('Frequency')
plt.show()
sns.histplot(x=datas['Age'],bins=10, kde=True, data=datas)
plt.show()
sns.histplot(x=datas['BMI'],bins=10, kde=True, data=datas)
plt.show()
sns.histplot(x=datas['pregs'],bins=10, kde=True, data=datas)
plt.show()
sns.histplot(x=datas['plas'],bins=10, kde=True, data=datas)
plt.show()
```
