```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
iris=pd.read_csv('Iris.csv')
iris
X, y = iris.iloc[:,:-1], iris.Species
```

```python
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_normalized)
```

```python
df = pd.DataFrame(data=X_pca, columns=["PC1", "PC2"])
df["label"] = y
df['label']=df['label'].map({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
df.to_csv("iris_dataset_2D", index=False)
```

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(df["PC1"], df["PC2"],
                      c=df["label"], cmap="viridis", edgecolor='k', s=50)
plt.colorbar(scatter, ticks=range(5), label="Digit Label")
plt.title("PCA of IRIS")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.show()
```
