```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import confusion_matrix
```

```python
df=pd.read_csv('iris_dataset_2D')
X_2D = df[["PC1", "PC2"]].values
y_true = df["label"].values
```

```python

agg_cluster = AgglomerativeClustering(n_clusters=3, linkage='ward')
cluster_labels = agg_cluster.fit_predict(X_2D)
plt.figure(figsize=(6, 5))
plt.scatter(df["PC1"], df["PC2"],c=cluster_labels, cmap="tab10", s=50)
plt.title("Agglomerative Clustering (n_clusters = 3, linkage = 'ward')")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.show() 
def purity_score(y_true, y_pred):
    con = confusion_matrix(y_true, y_pred)
    # Take the max value from each cluster (column-wise) and sum them
    return np.sum(np.amax(con, axis=0)) / np.sum(con)

# Calculate purity
purity = purity_score(y_true, cluster_labels)
print(f"Purity Score: {purity:.4f}")
```

```python
agg_cluster_complete = AgglomerativeClustering(n_clusters=3, linkage='complete')
cluster_labels_complete = agg_cluster_complete.fit_predict(X_2D)
plt.figure(figsize=(6, 5))
plt.scatter(df["PC1"], df["PC2"],c=cluster_labels_complete, cmap="tab10", s=50)
plt.title("Agglomerative Clustering (n_clusters = 3, linkage = 'comple')")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.show() 
purity = purity_score(y_true, cluster_labels_complete)
print(f"Purity Score: {purity:.4f}")
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
Z = linkage(X_2D, method='ward')
plt.figure(figsize=(8, 6))
dendrogram(Z, truncate_mode='lastp', p=30, color_threshold=10)
plt.axhline(y=6, c='red', ls='--')  # Threshold line
plt.title(f'Dendrogram ({method} linkage)')
plt.xlabel("Sample index or (cluster size)")
plt.ylabel("Distance")
plt.show()
print("optimal clusters=5")
```
